#### Introduction to Dictionaries
Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
# create a Dictionaries
empty_dict ={}
print(type(empty_dict))
```

```python

empty_dict=dict()
empty_dict
```

```python
student = {'name' : "Bicky", 'age' : 25, 'grade' : 'A'}
print(student)
print(type(student))
```

```python
#Single key is slways used
student = {'name' : "Bicky", 'age' : 25, 'age' : 26, 'grade' : 'A', 'name' : "Vikram"}
print(student)
```

```python
# Accessing elements in a dictionary
student = {'name' : "Bicky", 'age' : 25, 'grade' : 'A'}
print(student)
```

```python
## Accessing elements in a dictionary
print(student['name'])
print(student['age'])

## Accessing elements using get() method
print(student.get('grade'))
print(student.get('last_name'))
```

```python
## Modifying Dicitonary Elements
## Dictionaries are Mutable
print(student)
```

```python
student["age"] = 30
print(student)
```

```python
## add new key value pair
student["address"] = "habibhouar"
print(student)
```

```python
del student['grade'] ## delete key and value pair

print(student)
```

```python
## Dictionary Methods
keys = student.keys() ## get all keys
print(keys)
values = student.values() ## get all values
print(values)

items = student.items() ## get all items
print(items)
```

```python
## shallow copy
student_copy = student.copy()
print("Original Dictionary")
print(student)
print("Copied Dictionary")
print(student_copy)
```

```python
student['name'] = "Bikcy2"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy()
print(student_copy)
print(student)
```

```python
student['name'] = "Bikcy3"
print(student_copy1)
print(student)
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionatries, keys,values,or items

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Disctionaries
students={
    "student1":{"name":"Bikcy","age":32},
    "student2":{"name":"Shivam","age":35}
}
print(students)
```

```python
## Access nested dictionaries elementss
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python

students.items()
```

```python
## Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Practical Examples

## USe a dictionary to count he frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

#### Conclusion
Dictionaries are powerful tools in Python for managing key-value pairs. They are used in a variety of real-world scenarios, such as counting word frequency, grouping data, storing configuration settings, managing phonebooks, tracking inventory, and caching results.
