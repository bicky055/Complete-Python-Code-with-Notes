### Standard Library Overview

Python's Standard Library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.



```python
import array
arr = array.array('i', [1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(16))
```

```python
## Randon
import random
print(random.randint(1,10))
print(random.choice(['apple', 'Mango', 'Banana']))
```

```python
## File and Directory Access
import os
print(os.getcwd()) # Get Current Working Directory
```

```python
## High Level operations on files and colleection of files

import shutil
shutil.copy('source.txt', 'destination.txt')  # Copy file from source to destination
```

```python
## Data Serialization 
import json
data = {'name' : 'Bikcy', 'age': 26}

json_str = json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## CSV 
import csv

with open('example.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Alice', 30, 'New York'])
    writer.writerow(['Bob', 25, 'Los Angeles'])

with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)    
```

```python
# datetime 
from datetime import datetime, timedelta

now = datetime.now()
print("Current Date and Time:", now)

yesterday = now - timedelta(days = 1)
print("Yesterday's Date and Time:", yesterday)
```

```python
# time
import time
print(time.ctime())  # Current time in seconds since epoch
time.sleep(2)  # Sleep for 2 seconds    
print("Awake after 2 seconds")
print(time.ctime())  # Current time in a human-readable format
```

```python
# Regular Expressions
import re
pattern = r'\bfoo\b'
text = 'foo bar baz fooqux'
matches = re.search(pattern, text)
print(matches.group())
```

```python
pattern=r'\d+'
text='There are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
```
