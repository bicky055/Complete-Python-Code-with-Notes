# importing Modules in python : Modules and Packages

In python, modules and packages hepl organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(25))
print(sqrt(49))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4,5])
```

```python
from math import *
print(sqrt(36))
print(pi)
```

```python
from package.maths import addition
addition(5,10)
```
