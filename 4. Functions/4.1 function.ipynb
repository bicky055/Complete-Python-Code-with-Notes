##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
## Syntax
def function_name(parameterds):
    """Docsting"""
    #Function Body
    return expression
```

```python
# Why functions?
# 1. Reusability
# 2. Modularity
# 3. Readability
# 4. Maintainability

# Example
num = 24
if num%2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")    
```

```python
def is_even_or_odd(number):
    """This function find even or odd"""
    if number%2 == 0:
        return f"{number} is even"
    else:
        return f"{number} is odd"
```

```python
## Call this Function
is_even_or_odd(23)
```

```python
is_even_or_odd(12)
```

```python
# Function with Multiple parameters
def add(a,b):
    return a + b

result = add(2,3)
print(result)
```

```python
## Default parameters
def greet(name):
    print(f"Hello {name} Welcome to the paradise")

greet("Vikram")    
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome to the paradise")
greet("Bicky")    
```

```python
## Variable Length Arguments
## Positional and Keywords arguments

def print_numbers(*Bicky):
    for number in Bicky:
        print(number)
```

```python
print_numbers(1,2,3,4,5,"Bicky")
```

```python
# Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
        
```

```python
print_numbers(1,2,3,4,5, "Vikram")
```

```python
# Keywords Arguments
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Vikram", age=24, city="Bangalore")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f" positional argument: {val}")

    for key, value in kwargs.items():
        print(f"{key}: {value}")    
```

```python
print_details(1,2,3,4,5,"Bikcy", name="Vikram", age=24, city="Bangalore")
```

```python
# return statement
def multiply(a,b):
    return a*b
```

```python
multiply(2,5)
```

```python
## Return multiple parameters
def multiply(a,b):
    return a*b, a+b, a/b
```

```python
multiply(2,4)
```
