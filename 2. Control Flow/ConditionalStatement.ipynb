```python
## if statement
age = 25
if age>=18:
    print("you are allowed to vote in the elections")
    
```

```python
## else
age = 15
if age>=18:
    print("you are eligible for voting")
else:
    print("you are not eligible for voting")    
```

```python
# elif
age = 17
if age<13:
    print("you are a child")
elif age<18:
    print("you are a teenager")
else:
    print("you are an adult")    
```

```python
## Nested Conditional Statements
# you can place one or more if. elif, else satarements inside another if, elif or else statement.
# number is positive, negative or zero

num = int(input("Enter a number: "))
if num>0:
    print("The number is positive")
    if num%2 == 0:
        print("The number is even")
    else:
        print("The number is odd")    
else:
    if num == 0:
        print("The number is zero")
    else:
        print("The number is negative")
```

```python
# --- Practice Questions ---

## Determine if a year is a leap year or not
year = int(input("Enter a year: "))

if year%4 == 0:
    if year%100 == 0:
        if year%400 == 0:
            print(f"{year} is a leap year")
        else:
            print(f"{year} is not a leap year")
    else:
        print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")                
```

```python
# Assingment 
# simple calculator using if, elif and else statements

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

#print("Select operation:")

operation = input("Enter operation (+, -, *, /): ")

if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2        
    else:
        result = "Error! Division by Zero."
else:
    result = "Invalid operation"
print(f"The result of {num1} and {num2} with {operation} is : {result}")            

```

### Complex Example 3: Employee Bonus Calculation
   Calculate an employee's bonus based on their performance rating and years of service.

```python
# Employee Bonus Calculation

years_of_service = int(input("Enter years of service: "))
performance_rating = float(input("Enter performance rating (1.0 to 5.0): "))

# Determine bonus based percentage

if performance_rating >= 4.5:
    if years_of_service > 10:
        bonus_percentage = 20
    elif years_of_service >= 5:
        bonus_percentage = 15
    else:
        bonus_percentage = 10
elif performance_rating >= 3.5:
    if years_of_service > 10:
        bonus_percentage = 15
    elif years_of_service >= 5:
        bonus_percentage = 10
    else:
        bonus_percentage = 5
else:
    bonus_percentage = 0
# calculate bonus amount
salary = float(input("Enter your salary: "))
bonus_amount = (bonus_percentage / 100) * salary    

print(f"Your bonus amount is: ${bonus_amount:.2f}")
```
